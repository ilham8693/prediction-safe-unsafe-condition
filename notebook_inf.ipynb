# Model Inference

```python
# Import libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
```

```python
# Load model

model_imp = load_model('model.keras')
```

Model has been loaded.

- Data inference 1

```python
# Load data 

img1 = cv2.imread('image1.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

Read data inference or image1 to do prediction.

```python
# Preprocessing

# Resize as input model (img_height  img_width)
img1_resized = cv2.resize(img1, (200, 200))

# Scaling
img1_array = np.array(img1_resized) / 255.

# Change dimension
img1_dims = np.expand_dims(img1_array, axis=0)
```

Preproccesing needs to be done like dataset during modeling in preprocessing or feature engineering.

```python
# Prediction

prob1 = model_imp.predict(img1_dims)

# Convert to class (0 = safe, 1 = unsafe)
pred1 = 1 if prob1 >= 0.5 else 0
class_names = ['safe', 'unsafe']
label1 = class_names[pred1]

# Result
plt.imshow(img1)
plt.title(f"Prediction: {label1} (prob={prob1})")
plt.axis("off")
plt.show()
```

Based on prediction result, image1 is consider safe working condition with 5.8% probability. This is good because model can really tell it is safe, based on human vision, that is safety condition of working with complete proper protection.
